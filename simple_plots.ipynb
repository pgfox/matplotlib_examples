## Basic line plot

line styles that are available "-" , "--" , ":" (dotted line), "-." (dash dot line).
<br><br>

*Some* of the markers that are available: ".", ",", "o", "v", "^", "<", ">", "1","2","3"  ...
You can pass in latex code by specifying marker with " marker=r'$\Sigma$'"



```python
x_vals = np.arange(0,20,0.1)
print(x_vals)

# plot x vs sin(x)
plt.plot(x_vals, np.sin(x_vals),label="sin(x)",color="orange",linestyle='-')

#generate a second line on same graph, with marker and change the size 
plt.plot(x_vals,np.cos(x_vals),label="cos(x)",color="green" )


# mark every 5 points - to avoid a smear of markers
plt.plot(x_vals,np.cos(x_vals)+1,label="cos(x)+1",color="blue", marker='v',
         markevery=5, markersize=3)

# using latex to set the marker type.
plt.plot(x_vals,np.cos(x_vals)+3,label="cos(x)+3",color="indigo", marker='$\Sigma$',
         markevery=5, markersize=5, linestyle='None')


#axis limit set 
#set plot to display 0 and 20 on x axis
plt.xlim(0,20)
#set plot to display -4 and 4 on y axis
plt.ylim(-4,4)

# set labels on the graph
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()


```

### customising line 

Default line width is : linewidth=1 (short lw). <br><br>



Alpha is the transparency of the line being plotted. Alpha=0 completely transparent and alpha=1 is completely opaque. 


```python
x_vals = np.arange(0,20,0.1)
print(x_vals)

# plot x vs sin(x)
plt.plot(x_vals, np.sin(x_vals),label="sin(x)",color="blue",linestyle='-' , linewidth=0.5)

# plot x vs cos(x)
plt.plot(x_vals, np.cos(x_vals),label="sin(x)+2",color="green",linestyle='-' , 
         linewidth=1, alpha=0.3)
```

### order of x matters

Order of x array is important. The line will be drawn between points next to each other in the x-axis array 



```python
x_vals = np.arange(0,10,0.1)
print(x_vals)

# plot x vs sin(x)
plt.plot(x_vals, np.sin(x_vals),label="sin(x)",color="blue",linestyle='-' , linewidth=0.5)

#switch around the points in the x_vals array
for i in range(0,len(x_vals)):
    if (i % 6) == 0:
        temp = x_vals[(i-3)] 
        x_vals[(i-3)] = x_vals[i] 
        x_vals[i] = temp 
        
print(x_vals)
#plot mixed up points as line
plt.plot(x_vals, np.sin(x_vals)+2,label="mixed up",color="orange",linestyle='-' , linewidth=0.5) 

#plot mixed line as points
plt.plot(x_vals, np.sin(x_vals)+4,label="mixed up points",color="red",linestyle='none',
         marker=".",markersize=1,linewidth=0.5) 


plt.show()        
```

If you points are not ordered nicely than you should plot just markers rather than a line. They will still appear on the plot in the correct place( just not joined together)


## scatter plot

By **default** it will plot markers (with an edge) rather than a line.
<br><br>
Interesting - the scatter plot can color the points based no value
Useful for a 3rd dimension (for continuous values)
<br><br>
cmap can chnge the color palette used 
<br><br>
size is controlled by "s"

<br><br>
NOTE: Beware of the speed - for very large data the plot() method can be a significantly quicker than scatter().


```python
x_vals = np.arange(0,10,.1)

plt.scatter(x_vals,np.sin(x_vals))
plt.scatter(x_vals,np.sin(x_vals)+2, s=4, c=np.sin(x_vals), cmap="inferno", edgecolor="black")
plt.colorbar()
plt.scatter(x_vals,np.sin(x_vals)+4, c=np.sin(x_vals), edgecolor="none")
plt.colorbar()

```

you can also graduate the size based on continuous variables.
<br>
So in theory you could show 4 dimensions of data this way in a 2 d plain.
<li>x axis</li>
<li>y axis</li>
<li>colour</li>
<li>size</li>


```python

x_vals = np.arange(0,10,.1)
plt.scatter(x_vals,np.sin(x_vals)+2, s=np.power(x_vals,2), c=np.sin(x_vals), edgecolor="black")
plt.show()
```

## Histogram

The plt.histogram() is an easy way to bin up your data without having to manually choose bins

```python
values = np.random.normal(size=50)
print(f"randomly selected values from normal distribution {values}")
print(type(values))

plt.hist(values)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
